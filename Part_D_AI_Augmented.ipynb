## Step 3 & 4 — Testing Every Conversion

Each conversion is tested below. Results are marked:

### 🔢 int → other types

```python
# int → float ✅
print(float(5))         # Expected: 5.0
```

```python
# int → str ✅
print(str(42))          # Expected: "42"
print(type(str(42)))    # Expected: <class 'str'>
```

```python
# int → bool ✅
print(bool(0))          # Expected: False
print(bool(1))          # Expected: True
print(bool(-5))         # Expected: True  (any non-zero = True)
```

```python
# int → list ✅ (TypeError — int is not iterable)
try:
    print(list(5))
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
# int → tuple ✅ (TypeError — int is not iterable)
try:
    print(tuple(5))
except TypeError as e:
    print(f"TypeError: {e}")
```

### 🔣 float → other types

```python
# float → int ✅ (truncates, does NOT round)
print(int(3.99))        # Expected: 3  (not 4!)
print(int(-3.99))       # Expected: -3 (truncates toward zero)
```

```python
# float → str ✅
print(str(3.14))        # Expected: "3.14"
```

```python
# float → bool ✅
print(bool(0.0))        # Expected: False
print(bool(0.1))        # Expected: True
print(bool(-1.5))       # Expected: True
```

### 📝 str → other types

```python
# str → int ✅
print(int("42"))        # Expected: 42
```

```python
# str → int with float string ⚠️ AI MISSED THIS
# You CANNOT convert "3.14" directly to int — must go via float first
try:
    print(int("3.14"))      # ValueError!
except ValueError as e:
    print(f"ValueError: {e}")

# Correct way:
print(int(float("3.14")))   # Expected: 3
```

```python
# str → float ✅
print(float("3.14"))    # Expected: 3.14

# Invalid string → float ✅
try:
    print(float("hello"))
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
# str → bool ⚠️ AI MISSED THIS EDGE CASE
print(bool(""))         # Expected: False  (empty string)
print(bool("False"))    # Expected: True   (non-empty = ALWAYS True!)
print(bool("0"))        # Expected: True   ("0" is non-empty string!)
```

```python
# str → list ✅ (splits into individual characters)
print(list("abc"))      # Expected: ['a', 'b', 'c']
```

```python
# str → tuple ✅
print(tuple("abc"))     # Expected: ('a', 'b', 'c')
```

### ✔️ bool → other types

```python
# bool → int ✅
print(int(True))        # Expected: 1
print(int(False))       # Expected: 0
```

```python
# bool → float ✅
print(float(True))      # Expected: 1.0
print(float(False))     # Expected: 0.0
```

```python
# bool → str ✅
print(str(True))        # Expected: "True"  (capital T)
print(str(False))       # Expected: "False" (capital F)
```

### 📋 list & tuple conversions

```python
# list → tuple ✅
print(tuple([1, 2, 3]))     # Expected: (1, 2, 3)
```

```python
# list → str ⚠️ AI MISSED THIS
# str() gives repr form, NOT a clean joined string
print(str([1, 2, 3]))       # Expected: "[1, 2, 3]"  (includes brackets!)
# If you want clean: use join()
print(", ".join(["a", "b", "c"]))  # Expected: "a, b, c"
```

```python
# list → bool ✅
print(bool([]))             # Expected: False  (empty list)
print(bool([0]))            # Expected: True   (non-empty, even if contains 0!)
```

```python
# tuple → list ✅
print(list((1, 2, 3)))      # Expected: [1, 2, 3]
```

```python
# tuple → bool ✅
print(bool(()))             # Expected: False  (empty tuple)
print(bool((1,)))           # Expected: True
```

### ⚡ Bonus: float precision edge case ⚠️ AI MISSED THIS

```python
# IEEE 754 floating point — critical for financial calculations
print(0.1 + 0.2)            # Expected: 0.30000000000000004  (NOT 0.3!)
print(0.1 + 0.2 == 0.3)     # Expected: False

# Fix using decimal module:
from decimal import Decimal
print(Decimal("0.1") + Decimal("0.2"))   # Expected: 0.3 (exact!)
```

## Step 5 — Critical Evaluation (150–200 words)

The AI-generated type conversion matrix was largely accurate and covered the most common conversion pairs well. Standard conversions like `int→float`, `str→int`, `bool→int`, `list→tuple` and vice versa were all described correctly, making it a solid starting reference.

However, several important edge cases were missing or incomplete:

1. **`bool("False") → True`**: The AI failed to flag this dangerous trap. Any non-empty string is truthy — even `"False"` or `"0"`. This causes real bugs for beginners.

2. **`int("3.14") → ValueError`**: The AI did not mention that float-formatted strings cannot be converted directly to int. You must do `int(float("3.14"))` instead.

3. **`list(5) → TypeError`**: Some AI responses implied this might work like `range(5)`. It does not — integers are not iterable.

4. **`str([1,2,3]) → "[1, 2, 3]"`**: The AI did not clarify this gives a repr-style string with brackets, not a clean joined output.

5. **Float precision**: The `0.1 + 0.2 ≠ 0.3` issue was completely omitted — critical knowledge for anyone doing financial calculations.

**Overall AI accuracy: ~80%.** It is a useful starting point but must always be tested and verified. Blindly trusting AI output without running the code would introduce subtle, hard-to-find bugs.
