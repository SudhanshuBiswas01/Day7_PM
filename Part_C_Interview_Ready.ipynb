## Q1 — Conceptual: Data Types Deep Dive
**Task:** Predict the output of each line *without* running it first, then verify by executing.

### Line 1: `print(type(True))`
**Prediction:** `<class 'bool'>`  
**Reason:** `True` is a boolean literal. `type()` returns its class which is `bool`.

````python
print(type(True))
````

### Line 2: `print(isinstance(True, int))`
**Prediction:** `True`  
**Reason:** In Python, `bool` is a **subclass of `int`**. So `True` is technically an integer too. `isinstance()` checks the full inheritance chain.

````python
print(isinstance(True, int))
````

### Line 3: `print(True + True + False)`
**Prediction:** `2`  
**Reason:** `True = 1`, `False = 0` in integer arithmetic. So `1 + 1 + 0 = 2`.

````python
print(True + True + False)
````

### Line 4: `print(int(3.99))`
**Prediction:** `3`  
**Reason:** `int()` **truncates** toward zero — it does NOT round. So 3.99 becomes 3.

````python
print(int(3.99))
````

### Line 5: `print(bool("False"))`
**Prediction:** `True`  
**Reason:** ⚠️ Common trap! Any **non-empty string** is truthy in Python — even the string `"False"`. Only `""` (empty string) is falsy.

````python
print(bool("False"))
````

### Line 6: `print(bool(""))`
**Prediction:** `False`  
**Reason:** An **empty string** `""` has no content, so it is falsy in Python.

````python
print(bool(""))
````

### Line 7: `print(0.1 + 0.2 == 0.3)`
**Prediction:** `False`  
**Reason:** ⚠️ Floating-point precision issue (IEEE 754). `0.1 + 0.2` actually equals `0.30000000000000004` in binary, not exactly `0.3`. This is why financial apps use `decimal.Decimal` instead of `float`.

````python
print(0.1 + 0.2 == 0.3)
# See the actual value:
print(0.1 + 0.2)
````

### Line 8: `print("5" + "3")`
**Prediction:** `53`  
**Reason:** Both are **strings**, so `+` does **concatenation**, not addition. `"5" + "3"` = `"53"`.

````python
print("5" + "3")
````

### Line 9: `print(5 + 3)`
**Prediction:** `8`  
**Reason:** Both are **integers**, so `+` does normal arithmetic addition.

````python
print(5 + 3)
````

## Q2 — Coding: Build a Type Analyzer
Write a function `analyze_value(value)` that returns the value, its type, truthiness, and length (if applicable).

````python
def analyze_value(value):
    """
    Analyze a Python value and return its type, truthiness, and length.
    
    Args:
        value: Any Python value.
    
    Returns:
        A formatted string with value, type, truthiness, and length.
    """
    val_type = type(value).__name__
    truthy = bool(value)
    try:
        length = len(value)
    except TypeError:
        length = "N/A"
    return f"Value: {value} | Type: {val_type} | Truthy: {truthy} | Length: {length}"
````

````python
# Test all cases
print(analyze_value(42))
print(analyze_value(""))
print(analyze_value([1, 2, 3]))
print(analyze_value(True))
print(analyze_value((1, 2)))
print(analyze_value(3.14))
print(analyze_value(None))
````

## Q3 — Debug: Find & Fix 4 Bugs

### 🐛 Original Buggy Code
```python
name = input("Name: ")
age = input("Age: ")

if age >= 18:                             # BUG 1
    status = "Adult"
else:
    status = "Minor"

print(f"name is {age} years old and is a {status}")  # BUG 2
print(f"In 5 years: {age + 5}")          # BUG 3

score = 85.5
print(f"Score: {score:.0}")              # BUG 4
```

### 🔍 Bugs Identified

| # | Line | Bug | Fix |
|---|------|-----|-----|
| 1 | `if age >= 18` | `age` is a **string** (from `input()`), can't compare string to int → `TypeError` | Convert: `age = int(input(...))` |
| 2 | `f"name is {age}..."` | Prints the word `name` literally instead of the variable `{name}` | Change to `f"{name} is {age}..."` |
| 3 | `age + 5` | `age` is a string, `str + int` → `TypeError` | Fixed by Bug 1's fix (age is now int) |
| 4 | `{score:.0}` | Invalid f-string format spec — missing `f` → `ValueError` | Change to `{score:.0f}` |

````python
# ✅ CORRECTED VERSION

name = input("Name: ")
age = int(input("Age: "))          # FIX 1: convert str → int

if age >= 18:                      # Now works: int compared to int
    status = "Adult"
else:
    status = "Minor"

print(f"{name} is {age} years old and is a {status}")  # FIX 2: {name} not "name"
print(f"In 5 years: {age + 5}")   # FIX 3: works now since age is int

score = 85.5
print(f"Score: {score:.0f}")      # FIX 4: added 'f' to format spec
````

---
### ✅ Summary of All 4 Fixes

1. **`age = int(input(...))`** — `input()` always returns a string; must convert to `int` before comparison or arithmetic  
2. **`f"{name} is {age}..."`** — was printing the literal word `name` instead of the variable's value  
3. **`age + 5`** — fixed automatically once age became an `int` in Fix 1  
4. **`{score:.0f}`** — format spec needs the type character `f` (float); `.0` alone is invalid
